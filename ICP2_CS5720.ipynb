## Question 1: Full Name and Alternate Characters

```python
def fullname(first_name, last_name):
    return first_name + ' ' + last_name

def string_alternative(s):
    return s[::2]

first_name = input('Enter first name: ')
last_name = input('Enter last name: ')

full = fullname(first_name, last_name)
print('Full Name:', full)
print('Alternate Characters:', string_alternative(full))
```

## Question 2: Word Count from File

```python
word_count = {}

with open('input.txt', 'r') as f:
    for line in f:
        words = line.split()
        for word in words:
            word_count[word] = word_count.get(word, 0) + 1

with open('output.txt', 'w') as out:
    for word, count in word_count.items():
        out.write(f"{word}: {count}\n")

print('Word count written to output.txt')
```

## Question 3: Heights Conversion

```python
heights_in = [150, 155, 145, 148]

# Nested loop
heights_cm_loop = []
for h in heights_in:
    heights_cm_loop.append(h * 0.453592 * 2.54)

# List comprehension
heights_cm_comp = [h * 0.453592 * 2.54 for h in heights_in]

print(heights_cm_loop)
print(heights_cm_comp)
```

## Question 4: Employee Classes

```python
class Employee:
    count = 0

    def __init__(self, name, family, salary, department):
        self.name = name
        self.family = family
        self.salary = salary
        self.department = department
        Employee.count += 1

    @staticmethod
    def average_salary(employees):
        return sum(emp.salary for emp in employees) / len(employees)

class FulltimeEmployee(Employee):
    pass

e1 = Employee('John', 'Doe', 50000, 'HR')
e2 = FulltimeEmployee('Jane', 'Smith', 70000, 'IT')

print(Employee.average_salary([e1, e2]))
```
