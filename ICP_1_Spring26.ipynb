## 1(a) String operations

```python

chars = list(input("Enter a string: ").lower())

if len(chars) >= 2:
    del chars[0]
    del chars[-1]

chars.reverse()
print("".join(chars))
```

## 1(b) Arithmetic operations

```python

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)

if b != 0:
    print("Division:", a / b)
else:
    print("Division: Not possible")
```

## 2. Replace python with pythons

```python

sentence = input("Enter a sentence: ")
print(sentence.replace("python", "pythons"))
```

## 3. Letter Grade Program

```python

score = float(input("Enter class score: "))

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print("Grade:", grade)
```
